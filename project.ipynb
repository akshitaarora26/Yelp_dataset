**Table: Reviews**

Columns:
* review_id (VARCHAR): Unique identifier for each review.
* user_id (VARCHAR): Unique identifier for the user who left the review; links to Users table.
* business_id (VARCHAR): Unique identifier for the business being reviewed; links to Business table.
* stars (FLOAT): Star rating provided in the review, on a scale of 1.0 to 5.0.
* useful (INT): Count of "useful" votes this review received.
* funny (INT): Count of "funny" votes this review received.
* cool (INT): Count of "cool" votes this review received.
* text (TEXT): Full text of the review, providing content feedback.
* date (DATE): Date the review was written.

**Primary Key:** review_id

**Foreign Keys:**
* user_id references Users table (One-to-Many).
* business_id references Business table (One-to-Many).

Data Source: Extracted from yelp_academic_dataset_review.json JSON file.

1. FIRST QUERY: Analyse the categories which do not have many restaurants

```python
query = """
SELECT categories,count(*) as countOfCategories
FROM business
WHERE categories LIKE '%Restaurant%' and is_open=1
group by categories order by countOfCategories asc ;
"""

# Execute the query and store the results in a DataFrame
df_popular_categories = pd.read_sql(query, engine)
print(df_popular_categories)



```

Here , we can see categories of Acai Bowls are very less. Let's explore this category.

2. SECOND QUERY: Analyse the count of Acai Bowls category in each city

```python
query = """
SELECT city, COUNT(*) AS cnt
FROM business
WHERE categories LIKE 'Acai Bowls%'
GROUP BY city
ORDER BY cnt ASC
"""

# Execute the query and store the results in a DataFrame
df_popular_loc = pd.read_sql(query, engine)
print(df_popular_loc)
```

Acai restaurants are highest in Tampa. Let's pick St.Louis
